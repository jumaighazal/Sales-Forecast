```python
train_data = train_data.join(items, on='item_id', rsuffix='_').join(shops, on='shop_id', rsuffix='_').join(item_categories, on='item_category_id', rsuffix='_').drop(['item_id_', 'shop_id_', 'item_category_id_'], axis=1)

```

```python
print('train_data rows: ', train_data.shape[0])
print('train_data columns: ', train_data.shape[1])
```

```python
train_data = train_data.query('item_price > 0')
```

```python
#Useful features
train_data = train_data[['date', 'date_block_num', 'shop_id', 'item_category_id', 'item_id', 'item_price', 'item_cnt_day']]
```

```python
train_data['date'] = pd.to_datetime(train_data['date'], format='%d.%m.%Y')
```

```python
train_data['month'] = train_data['date'].dt.month
train_data['day'] = train_data['date'].dt.day
train_data['year'] = train_data['date'].dt.year
train_data['weekday'] = train_data['date'].dt.weekday
```

```python
train_data['revenue'] = train_data.item_price * train_data.item_cnt_day
```

```python
#Average Item price
item_price = train_data[['item_id', 'item_price']].groupby('item_id')['item_price'].mean()

item_price = item_price.reset_index()
item_price = item_price.rename(columns={'item_price': 'avg_item_price'})

item_price.head(5)
```

```python
df = train_data[['date_block_num', 'item_id']].groupby(by='date_block_num', as_index=False).nunique()

fig, ax = plt.subplots(figsize=(16,4))
plt.plot(df.item_id, sns.xkcd_rgb["denim blue"], lw=4)
plt.title("Count of Unique items sold")

df = train_data[['date_block_num', "item_cnt_day"]].groupby(by='date_block_num', as_index=False).sum()

f, ax = plt.subplots(figsize=(16, 4))
plt.plot(df.item_cnt_day, sns.xkcd_rgb["denim blue"], lw=4)
plt.title("Sum of items sold")

df = train_data[['date_block_num', 'revenue']].groupby(by='date_block_num', as_index=False).sum()

f, ax = plt.subplots(figsize=(16, 4))
plt.plot(df.revenue, sns.xkcd_rgb["denim blue"], lw=4)
plt.title("Total revenue generated")
```
